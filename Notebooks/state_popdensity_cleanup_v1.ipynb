```python
import pandas as pd
# Import and read the data
state_df = pd.read_csv("../Resources/state_pop_per_10yrs.csv")
state_df.head(10)
```

```python
state_df.columns
```

```python
state_df.drop(['Resident Population Density Rank','Number of Representatives',
               'Change in Number of Representatives','Change in Number of Representatives', 
              'Average Apportionment Population Per Representative',
              'Unnamed: 11','Unnamed: 12','Unnamed: 13'], axis=1, inplace=True)
state_df.head()
```

```python
state_df.rename(columns={'State_Full':'State Full'}, inplace=True)
state_df.head()
```

```python
state_df.drop(['Resident Population Density'], axis=1, inplace=True)
state_df.head()
```

```python
state_df = state_df.set_index(['Geography Type'])
state_df.head()
```

```python
state_df.drop(index=['Nation', 'Region'], inplace=True)
```
