```python
def function (cls):
    fire_data2_df=fire_data_df.loc[fire_data_df["Fire_Size_Class"]==cls]
    cls_y = np.reshape(fire_data2_df.Fire_Size_Class.values,(-1,1))
    fire_clean_df=fire_data2_df.drop(columns=["Fire_Size_Class","Discovery_Month","Discovery_Year",
                                          "Fire_Magnitude",
                                           "Temp_Pre_30","Temp_Pre_15","Temp_Pre_7",
                                           "Wind_Pre_30","Wind_Pre_15","Wind_Pre_7",
                                           "Hum_Pre_30", "Hum_Pre_15", "Hum_Pre_7",
                                           "Prec_Pre_30", "Prec_Pre_15","Prec_Pre_7",
                                           "Vegetation","Fire_Size"
                                           ])
    fire_cat = fire_clean_df.dtypes[fire_clean_df.dtypes == "object"].index.tolist()
    fire_clean_df[fire_cat].nunique()
    enc = OneHotEncoder(sparse=False)
    encode_df = pd.DataFrame(enc.fit_transform(fire_clean_df[fire_cat]),index=fire_clean_df[fire_cat].index)
    encode_df.columns = enc.get_feature_names(fire_cat)
    x_df = fire_clean_df.merge(encode_df,left_index=True, right_index=True)
    cls_df = x_df.drop(fire_cat,1)
    return cls_df, cls_y

#test_df=pd.concat([classd_x,classe_x,])


#classd_x,classd_Y =function ("D")
```

```python
class_b_X, class_b_y= function ("B")
class_c_X, class_c_y= function ("C")
class_d_X, class_d_y= function ("D")
class_e_X, class_e_y= function ("E")
class_f_X, class_f_y= function ("F")
class_g_X, class_g_y= function ("G")

# Split training/test datasets
X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(class_b_X, class_b_y, random_state=1)
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(class_c_X, class_c_y, random_state=1)
X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(class_d_X, class_d_y, random_state=1)
X_train_e, X_test_e, y_train_e, y_test_e = train_test_split(class_e_X, class_e_y, random_state=1)
X_train_f, X_test_f, y_train_f, y_test_f = train_test_split(class_f_X, class_f_y, random_state=1)
X_train_g, X_test_g, y_train_g, y_test_g = train_test_split(class_g_X, class_g_y, random_state=1)
```
