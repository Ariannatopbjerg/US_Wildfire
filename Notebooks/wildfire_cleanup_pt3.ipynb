```python
import pandas as pd
# Import our input dataset
fire_df = pd.read_csv('../Resources/wildfire_v2_clean_data.csv')
fire_df.head()
```

```python
fire_df.drop(['remoteness'], axis=1, inplace=True)
```

```python
fire_df.rename(columns={'fire_size':'Fire_Size', 
                        'fire_size_class':'Fire_Size_Class', 
                        'fire_cause':'Fire_Cause', 
                        'fire_size_class':'Fire_Size_Class', 
                        'latitude':'Latitude', 
                        'longitude':'Longitude', 
                        'state':'State',
                        'discovery_month':'Discovery_Month',
                        'discovery_date':'Discovery_Date',
                        'discovery_year':'Discovery_Year',
                        'fire_mag':'Fire_Magnitude',
                        'Temp_pre_30':'Temp_Pre_30',
                        'Temp_pre_15':'Temp_Pre_15',
                        'Temp_pre_7':'Temp_Pre_7',
                        'Temp_cont':'Temp_Present',
                        'Wind_pre_30':'Wind_Pre_30',
                        'Wind_pre_15':'Wind_Pre_15',
                        'Wind_pre_7':'Wind_Pre_7',
                        'Wind_cont':'Wind_Present',
                        'Hum_pre_30':'Hum_Pre_30',
                        'Hum_pre_15':'Hum_Pre_15',
                        'Hum_pre_7':'Hum_Pre_7',
                        'Hum_cont':'Hum_Present',
                        'Prec_pre_30':'Prec_Pre_30',
                        'Prec_pre_15':'Prec_Pre_15',
                        'Prec_pre_7':'Prec_Pre_7',
                        'Prec_cont':'Prec_Present'}, inplace=True)
fire_df.head()
```

```python
fire_clean_df1=fire_df.loc[(fire_df["Temp_Pre_30"]!=0)&
                         (fire_df["Temp_Pre_15"]!=0)&
                         (fire_df["Temp_Pre_7"]!=0)&
                         (fire_df["Temp_Present"]!=0)&
                         (fire_df["Wind_Pre_30"]!=0)&
                         (fire_df["Wind_Pre_15"]!=0)&
                         (fire_df["Wind_Pre_7"]!=0)&
                         (fire_df["Wind_Present"]!=0)&
                         (fire_df["Hum_Pre_30"]!=0)&
                         (fire_df["Hum_Pre_15"]!=0)&
                         (fire_df["Hum_Pre_7"]!=0)&
                         (fire_df["Hum_Present"]!=0)&
                         (fire_df["Prec_Pre_30"]!=0)&
                          (fire_df["Prec_Pre_15"]!=0)&
                          (fire_df["Prec_Pre_7"]!=0)&
                          (fire_df["Prec_Present"]!=0)]
```

```python
fire_clean_df2=fire_clean_df1.loc[(fire_df["Temp_Pre_30"]!=-1)&
                         (fire_df["Temp_Pre_15"]!=-1)&
                         (fire_df["Temp_Pre_7"]!=-1)&
                         (fire_df["Temp_Present"]!=-1)&
                         (fire_df["Wind_Pre_30"]!=-1)&
                         (fire_df["Wind_Pre_15"]!=-1)&
                         (fire_df["Wind_Pre_7"]!=-1)&
                         (fire_df["Wind_Present"]!=-1)&
                         (fire_df["Hum_Pre_30"]!=-1)&
                         (fire_df["Hum_Pre_15"]!=-1)&
                         (fire_df["Hum_Pre_7"]!=-1)&
                         (fire_df["Hum_Present"]!=-1)&
                         (fire_df["Prec_Pre_30"]!=-1)&
                          (fire_df["Prec_Pre_15"]!=-1)&
                          (fire_df["Prec_Pre_7"]!=-1)&
                          (fire_df["Prec_Present"]!=-1)]
fire_clean_df2
```

```python
fire_clean_df2.to_csv('../Resources/wildfire_data_final.csv', index=False)
```
